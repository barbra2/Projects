## 1. Пример импорта данных. Грузим данные

```python
import pandas as pd
path = "NY-House-Dataset.csv"
data = pd.read_csv(path, sep=',')  # если нужно, можно поменять разделитель

data
```

## 2. Понимаем, какая перед нами задача

У меня задача регресии, т.к. цена на аппартаменты может быть любым числом -$\infty$ до $\infty$.

(взято из 2й лабораторной)

### Датасет состоит из следующих переменных
- *BROKERTITLE*: Название брокера
- *TYPE*: Тип недвижимости
- *PRICE*: Цена недвижимости
- *BEDS*: Количество спален
- *BATH*: Количество ванных комнат
- *PROPERTYSQFT*: Площадь недвижимости в квадратных футах
- *ADDRESS*: Адрес
- *STATE*: Штат

- *ADMINISTRATIVE_AREA_LEVEL_2*: Административный район
- *LOCALITY*: Населенный пункт
- *SUBLOCALITY*: Подрайон
- *STREET_NAME*: Название улицы
- *LONG_NAME*: Длинное название

- *LONGITUDE*: Долгота

Моя целевая переменная(y) - price. Все остальные переменные - признаки(х)

## 3. Делаем предобработку данных

### Разделение выборки на тренировочную (train) и тестовую (test)

```python
from sklearn.model_selection import train_test_split
import numpy as np

data['PRICE'] = np.log1p(data['PRICE'])

x = data.drop(columns=['PRICE'])
y = data['PRICE']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=17)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

### Проверка пропусков в данных (выкидываю наблюдения с прощенными значениями)

```python
print(data.isnull().sum())
```

```python
#data.dropna(how = 'any')
```

### Нормировка численных переменных

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

numeral = ['BEDS', 'BATH', 'PROPERTYSQFT', 'LATITUDE', 'LONGITUDE']
scaler = StandardScaler()
x_train[numeral] = x_train[numeral].astype(float)
x_test[numeral] = x_test[numeral].astype(float)  

x_train.loc[:, numeral] = scaler.fit_transform(x_train[numeral])
x_test.loc[:, numeral] = scaler.transform(x_test[numeral])
```

### Категориальные признаки

```python
x_train = x_train.drop(columns=['BROKERTITLE', 'ADDRESS', 'MAIN_ADDRESS', 'ADMINISTRATIVE_AREA_LEVEL_2', 'STREET_NAME', 'LONG_NAME', 'FORMATTED_ADDRESS'])
x_test = x_test.drop(columns=['BROKERTITLE', 'ADDRESS', 'MAIN_ADDRESS', 'ADMINISTRATIVE_AREA_LEVEL_2', 'STREET_NAME', 'LONG_NAME', 'FORMATTED_ADDRESS'])
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['TYPE', 'STATE', 'LOCALITY', 'SUBLOCALITY']
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

X_train_encoded = pd.DataFrame(
    encoder.fit_transform(x_train[categorical_cols]),
    index=x_train.index, 
    columns=encoder.get_feature_names_out(categorical_cols)  
)

X_test_encoded = pd.DataFrame(
    encoder.transform(x_test[categorical_cols]),
    index=x_test.index,
    columns=encoder.get_feature_names_out(categorical_cols)
)

x_train = x_train.drop(columns=categorical_cols).join(X_train_encoded)
x_test = x_test.drop(columns=categorical_cols).join(X_test_encoded)
```

## Обучение на тренировочном множестве

Добавим полиномиальные признаки, потому что по результатам предыдущих обучений, линейная модель недообучается

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree= 2, include_bias=False)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

### Линейная модель

```python
from sklearn.linear_model import LinearRegression 

lr = LinearRegression()
lr.fit(x_train_poly, y_train)

y_train_pred_lr = lr.predict(x_train_poly)
y_test_pred_lr = lr.predict(x_test_poly)
```

### Деревянная модель

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor(max_depth=10)
tree.fit(x_train, y_train)

y_train_pred_tree = tree.predict(x_train)
y_test_pred_tree = tree.predict(x_test)
```

### k-ближайших соседей

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train, y_train)

y_train_pred_knn = knn.predict(x_train)
y_test_pred_knn = knn.predict(x_test)
```

### Случайный лес

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=100, max_depth=10)
rfr.fit(x_train, y_train)

y_train_pred_rfr = rfr.predict(x_train)
y_test_pred_rfr = rfr.predict(x_test)
```

### Подсчёт метрик на train и test множествах

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import numpy as np


mae_lr_train = mean_absolute_error(y_train, y_train_pred_lr)
mae_lr_test = mean_absolute_error(y_test, y_test_pred_lr)
rmse_lr_train = np.sqrt(mean_squared_error(y_train, y_train_pred_lr))
rmse_lr_test = np.sqrt(mean_squared_error(y_test, y_test_pred_lr))
mape_lr_train = mean_absolute_percentage_error(y_train, y_train_pred_lr)
mape_lr_test = mean_absolute_percentage_error(y_test, y_test_pred_lr)


mae_dt_train = mean_absolute_error(y_train, y_train_pred_tree)
mae_dt_test = mean_absolute_error(y_test, y_test_pred_tree)
rmse_dt_train = np.sqrt(mean_squared_error(y_train, y_train_pred_tree))
rmse_dt_test = np.sqrt(mean_squared_error(y_test, y_test_pred_tree))
mape_dt_train = mean_absolute_percentage_error(y_train, y_train_pred_tree)
mape_dt_test = mean_absolute_percentage_error(y_test, y_test_pred_tree)


mae_knn_train = mean_absolute_error(y_train, y_train_pred_knn)
mae_knn_test = mean_absolute_error(y_test, y_test_pred_knn)
rmse_knn_train = np.sqrt(mean_squared_error(y_train, y_train_pred_knn))
rmse_knn_test = np.sqrt(mean_squared_error(y_test, y_test_pred_knn))
mape_knn_train = mean_absolute_percentage_error(y_train, y_train_pred_knn)
mape_knn_test = mean_absolute_percentage_error(y_test, y_test_pred_knn)


mae_rf_train = mean_absolute_error(y_train, y_train_pred_rfr)
mae_rf_test = mean_absolute_error(y_test, y_test_pred_rfr)
rmse_rf_train = np.sqrt(mean_squared_error(y_train, y_train_pred_rfr))
rmse_rf_test = np.sqrt(mean_squared_error(y_test, y_test_pred_rfr))
mape_rf_train = mean_absolute_percentage_error(y_train, y_train_pred_rfr)
mape_rf_test = mean_absolute_percentage_error(y_test, y_test_pred_rfr)


print("Линейная регрессия:")
print(f"MAE (train): {mae_lr_train:.4f}, MAE (test): {mae_lr_test:.4f}")
print(f"RMSE (train): {rmse_lr_train:.4f}, RMSE (test): {rmse_lr_test:.4f}")
print(f"MAPE (train): {mape_lr_train:.4f}, MAPE (test): {mape_lr_test:.4f}")

print("\nДеревья решений:")
print(f"MAE (train): {mae_dt_train:.4f}, MAE (test): {mae_dt_test:.4f}")
print(f"RMSE (train): {rmse_dt_train:.4f}, RMSE (test): {rmse_dt_test:.4f}")
print(f"MAPE (train): {mape_dt_train:.4f}, MAPE (test): {mape_dt_test:.4f}")

print("\nK-ближайшие соседи:")
print(f"MAE (train): {mae_knn_train:.4f}, MAE (test): {mae_knn_test:.4f}")
print(f"RMSE (train): {rmse_knn_train:.4f}, RMSE (test): {rmse_knn_test:.4f}")
print(f"MAPE (train): {mape_knn_train:.4f}, MAPE (test): {mape_knn_test:.4f}")

print("\nСлучайный лес:")
print(f"MAE (train): {mae_rf_train:.4f}, MAE (test): {mae_rf_test:.4f}")
print(f"RMSE (train): {rmse_rf_train:.4f}, RMSE (test): {rmse_rf_test:.4f}")
print(f"MAPE (train): {mape_rf_train:.4f}, MAPE (test): {mape_rf_test:.4f}")
```

 1. Какая модель справилась лучше с поставленной задачей?
    Случайный лес имеет лучшие метрики (значения MAE, RMSE, MAPE меньше чем у других моделей)
2. Имеет ли место переобучение?
    Да, в линейной регрессии, тк ошибка на train существенно ниже чем для test. Так как метрики на обучающем наборе данных лучше, чем на тестовом, модель запомнила данные, но не смогла их обощить
3. Имеет ли место недообучение?
    Недообучение было в линейной регрессии, полиномиальные признаки увеличили сложность модели, но модель не справляется с обощением данных, поэтому вместо недообучения всплыло переобучение 
4. Как можно улучшить метрики моделей?
    1) Можно применить регуляризацию 
    2) Настроить гиперпараметры
