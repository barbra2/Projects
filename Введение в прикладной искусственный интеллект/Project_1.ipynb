## 7. Гистограммы 5 любых признаков

```python
categories = ['BEDS', 'BATH', 'PROPERTYSQFT', 'LATITUDE', 'LONGITUDE']
plt.figure(figsize=(15, 10))
for i, category in enumerate(categories, 1):
    plt.subplot(2, 3, i)
    data[category].hist(bins=50, edgecolor='black') 
    plt.title(f'Распределение {category}')
plt.tight_layout()
plt.show()
```

1. На первых двух графиках можно заметить моду около значений 1-3
2. На графике распределения площади, можно заметить тенденцию на малую площадь
3. График распределения широты показывает, что большая часть объектов находятся на широтах от 40.7 до 40.8 
4. На графике распределения долготы  можно заметить, что большая часть объектов нахоодится на долготе от -74.0 до -73.9, в то время как на некоторых широтах в диспазоне от -74.1 до -74 вообще нет никаких объектов.


## 8.

```python
import seaborn as sns
plt.figure(figsize=(15, 10))
categories = ['BEDS', 'BATH', 'PROPERTYSQFT', 'LATITUDE', 'LONGITUDE']
for i, category in enumerate(categories, 1):
    plt.scatter(data[category], data['PRICE'])
    sns.jointplot(x=data[category], y=data['PRICE'], kind="hex", color="#4CB391")
    plt.show()

```

1. На графике зависимости цены от площади можно заметить, что большая часть значений находится в начале оси OX
2. На графиках зависимости цены от широты и долготы можно заметить, что цена выше если объект находится на широтах от 40.7 до 40.8 и в диапазоне долготы от -74 до -73.9
